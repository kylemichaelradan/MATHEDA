#  Normal Distribution

```python
mean = df['Current'].mean()
std = df['Current'].std(ddof=1)

x_min = df['Current'].min()
x_max = df['Current'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
         color = '#00ff00',
         linewidth = 2)

plt.title('Normal Distribution of Current Values',
          font = 'Times New Roman',
          fontsize = 12)
plt.xlabel('Current (mA)')
plt.ylabel('Probability Density')
plt.show()
```

```python
plt.hist(df['Current'], bins = 5,
         density = 'true',
         alpha = 0.8,
         color = '#00ff00',
         edgecolor = '#000000')
plt.plot(x, p,
         color = "#0067ff",
         linewidth = 2)

plt.title('Normal Distribution of Current Values',
          font = 'Times New Roman',
          fontsize = 12)
plt.xlabel('Current (mA)')
plt.ylabel('Probability Density')

plt.show()
```

# Standard Normal Distribution N(0, 1) 

```python
mean = df['Z-score'].mean()
std = df['Z-score'].std(ddof=1)

x_min = df['Z-score'].min()
x_max = df['Z-score'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
         color = '#00ff00',
         linewidth = 2)

plt.title('Normal Distribution of Current Values',
          font = 'Times New Roman',
          fontsize = 12)
plt.xlabel('Z-Score')
plt.ylabel('Probability Density')
plt.show()
```
